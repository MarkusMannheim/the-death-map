```python
import pandas as pd

death_stats = pd.read_csv("deaths.csv")
death_stats = death_stats[(death_stats["Year"] == "2015-2017") & (death_stats["Rate"] != "NP") & (death_stats["Sex"] == "Persons")]
death_stats = death_stats[["SA3 code", "SA3 name", "Rate"]]
death_stats.set_index("SA3 name", inplace=True)
death_stats["Rate"] = death_stats["Rate"].astype("float")
death_stats
```

```python
edu_stats = pd.read_csv("sa3AgeEducation.csv", index_col="region")
edu_stats = edu_stats[["uni"]]
edu_stats = edu_stats.sort_values("uni", ascending=False)
edu_stats
```

```python
def get_state(code):
    states = ["NSW", "Vic", "Qld", "SA", "WA", "Tas", "NT", "ACT"]
    id = int(code[:1]) - 1
    return states[id]
data = death_stats.merge(edu_stats, left_index=True, right_index=True)
data = data[["SA3 code", "Rate", "uni"]]
data.columns = ["code", "rate", "uni"]
data["state"] = data["code"].apply(get_state)
data.index = data.index + " (" + data["state"] + ")"
data = data[["rate", "uni"]]
data.to_csv("education.csv", index_label="region")
data
```
