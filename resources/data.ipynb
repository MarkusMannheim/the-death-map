```python
import pandas as pd
data = pd.read_csv("deathStats.csv", na_values=["—", "np"], thousands=",")

def header_check(x):
    start = x.rfind("(") + 1
    string = x[start:-1]
    if string.find("-") > -1:
        return True
    else:
        return False
    
data["header"] = data["cause"].apply(header_check)
data_raw = data[data["header"] == False]
data_raw
```

```python
deaths_all = data_raw["people"].sum()
deaths_all
```

```python
deaths_avoidable = data_raw[data_raw["avoidable"] == "x"]
deaths_avoidable = deaths_avoidable.sort_values("Persons", ascending=False)
deaths_avoidable
```

```python
data_headers = data[data["header"] == True]
data_headers = data_headers.sort_values("Persons", ascending=False)
data_headers
```

```python
data_all = data[data["avoidable"] == "x"].sort_values("Persons", ascending=False)
data_all.head(25)
data_all.to_csv("data_all.csv", index=False)
```
